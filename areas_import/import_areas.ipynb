```python
world_areas = './static-areas.json'
```

```python
from datetime import datetime, timezone
import json
import requests

with open(world_areas) as f:
  areas = json.load(f)

  names = set()
  succeeded = 0
  failed = 0
  duplicated = 0
  for i, area in enumerate(areas):
    if area['name'] in names:
      duplicated += 1
      continue
    names.add(area['name'])
    contact_email = ''
    if 'contact_email' in area:
      contact_email = area['contact_email']
    elif 'contactEmail' in area:
      contact_email = area['contactEmail']
      del area['contactEmail']
    elif 'coordinates' in area and 'properties' in area['coordinates']:
      props = area['coordinates']['properties']
      if 'email' in props:
        contact_email = props['email']
      elif 'contact:email' in props:
        contact_email = props['contact:email']
    area['contact_emails'] = [
      {
        'email': contact_email,
        'consent_report': True,
      }
    ]
    if 'contactName' in area:
      contact_name = area['contactName']
      del area['contactName']
      area['contact_name'] = contact_name
    
    create_date = datetime.now(timezone.utc)
    area['created_at'] = create_date.strftime('%Y-%m-%dT%H:%M:%S.%fZ')
    area['updated_at'] = create_date.strftime('%Y-%m-%dT%H:%M:%S.%fZ')
    
    resp = requests.post('http://dev.api.cleanapp.io:8080/create_or_update_area', json={
      'version': '2.0',
      'area': area,
    })
    if resp.status_code == 200:
      succeeded += 1
    else:
      failed += 1
    if i % 10 == 0:
      print('Imported', i+1, 'areas, succeeded:', succeeded, 'duplicated:', duplicated, 'failed:', failed)
```
